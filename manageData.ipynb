```python
numeric_cols = ['RAPESEED(EUR/TON)', 'RAPESEED OIL(EUR/TON', 'SOYBEAN(Usc/Bsh)', 'SOYBEAN OIL(EUR/TON)',
                'USED COOKING OIL(EUR/TON)', 'BIODIESEL DE SOYBEAN OIL(USD/TON)', 'BIODIESEL DE RAPESEED OIL(USD/TON)',
                'BIODIESEL DE USED COOKING OIL(USD/TON)', 'GASOLEO(USD/TON)', 'GASOLINA(USD/TON)', 'CAMBIO EUR USD',
                'ELETRICIDADE', 'GAS NATURAL', 'CO2 EMISSIONS', 'COAL']

# Replace commas with dots, convert to numeric, and handle missing values
data[numeric_cols] = data[numeric_cols].replace(',', '.', regex=True).apply(pd.to_numeric, errors='coerce')

# Print the updated data types
print(data.dtypes)
```

```python
data

```

```python
def convert_usd_to_euro_TON(df, from_column,to_column, exchange_rate_column):
    print(from_column)
    df[to_column] = df[from_column] * df[exchange_rate_column]

    # Drop the original 'from_column'
    df = df.drop(columns=[from_column])
    return df
```

```python
columns_to_convert = {
    "BIODIESEL DE SOYBEAN OIL(USD/TON)": "BIODIESEL DE SOYBEAN OIL(EUR/TON)",
    "BIODIESEL DE USED COOKING OIL(USD/TON)": "BIODIESEL DE USED COOKING OIL(EUR/TON)",
    "BIODIESEL DE RAPESEED OIL(USD/TON)" : "BIODIESEL DE RAPESEED OIL(EUR/TON)",
    "GASOLEO(USD/TON)" : "GASOLEO(EUR/TON)",
    "GASOLINA(USD/TON)" : "GASOLINA(EUR/TON)"
}

for from_column, to_column in columns_to_convert.items():
    data = convert_usd_to_euro_TON(data, from_column, to_column, "CAMBIO EUR USD")
```

```python
data
```

```python
def convert_UscBsh_to_USDTON(df, from_column,to_column):
    df[to_column] = (df[from_column] * 36.7437019880332)/100
    df = df.drop(columns=[from_column])
    return df
```

```python
data = convert_UscBsh_to_USDTON(data, "SOYBEAN(Usc/Bsh)", "SOYBEAN(USD/TON)")
data = convert_usd_to_euro_TON(data, "SOYBEAN(USD/TON)", "SOYBEAN(EUR/TON)", "CAMBIO EUR USD")
```
